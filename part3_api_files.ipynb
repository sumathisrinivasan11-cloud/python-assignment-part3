```python
# ---- TASK1 Part B: Read from File ----

# Step 1: Read all lines
with open(text_path, "r", encoding="utf-8") as file:
    lines = file.readlines()


# Step 2: Print numbered lines (strip newline)
print("\n=== Numbered Notes ===")
for i, line in enumerate(lines, start=1):
    print(f"{i}. {line.strip()}")

# Step 3: Count total lines
print("\nTotal number of lines:", len(lines))

# Step 4: Keyword search (case-insensitive)
keyword = input("\nEnter a keyword to search: ").lower()

matches = []
for line in lines:
    if keyword in line.lower():
        matches.append(line.strip())

# Step 5: Print results
if matches:
    print("\nMatching lines:")
    for m in matches:
        print("-", m)
else:
    print("No lines found containing that keyword.")
```

```python
# --- Task 2 — API Integration ----
import requests

BASE_URL = "https://dummyjson.com/products"

# Step 1: Fetch 20 products
response = requests.get(f"{BASE_URL}?limit=20")
data = response.json()

products = data["products"]

# Print table header
print("ID  | Title                          | Category        | Price     | Rating")
print("----|--------------------------------|-----------------|-----------|--------")

for p in products:
    print(f"{p['id']:<4}| {p['title']:<30}| {p['category']:<15}| ${p['price']:<9}| {p['rating']}")
```

```python
# Filter products with rating >= 4.5
filtered = [p for p in products if p["rating"] >= 4.5]

# Sort by price (descending)
sorted_filtered = sorted(filtered, key=lambda x: x["price"], reverse=True)

print("\n=== Filtered (rating ≥ 4.5) & Sorted by Price (desc) ===")
for p in sorted_filtered:
    print(f"{p['title']} — ${p['price']} — Rating: {p['rating']}")
```

```python
# Fetch laptops category
laptop_response = requests.get(f"{BASE_URL}/category/laptops")
laptop_data = laptop_response.json()

print("\n=== Laptops Category ===")
for p in laptop_data["products"]:
    print(f"{p['title']} — ${p['price']}")
```

```python
# Task 3 — Exception Handling
# Part A — Guarded Calculator
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Error: Cannot divide by zero"
    except TypeError:
        return "Error: Invalid input types"


# Testing
print(safe_divide(10, 2))     # valid
print(safe_divide(10, 0))     # division by zero
print(safe_divide("ten", 2))  # invalid type
```

```python
# Part B — Guarded File Reader

def read_file_safe(filename):
    try:
        with open(filename, "r", encoding="utf-8") as file:
            return file.read()
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
    finally:
        print("File operation attempt complete.")

print(read_file_safe("MyFiles/python_notes.txt"))   # should succeed
print(read_file_safe("ghost_file.txt"))     # should fail gracefully
```

```python
# Part C — Robust API Calls (with try–except)

import requests

def safe_get(url):
    try:
        response = requests.get(url, timeout=5)
        return response
    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print("Unexpected error:", e)


def safe_post(url, json_data):
    try:
        response = requests.post(url, json=json_data, timeout=5)
        return response
    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print("Unexpected error:", e)
```

```python
# Usage Example (from Task 2)

BASE = "https://dummyjson.com/products"

resp = safe_get(f"{BASE}?limit=20")
if resp and resp.status_code == 200:
    print("Fetched products successfully.")
```

```python
# Part D — Input Validation Loop (Product Lookup)

BASE = "https://dummyjson.com/products"

while True:
    user_input = input("Enter a product ID to look up (1–100), or 'quit' to exit: ")

    if user_input.lower() == "quit":
        print("Exiting lookup.")
        break

    # Validate integer
    if not user_input.isdigit():
        print("Invalid input. Please enter a number between 1 and 100.")
        continue

    product_id = int(user_input)

    # Validate range
    if not (1 <= product_id <= 100):
        print("ID must be between 1 and 100.")
        continue

    # Make API call
    response = safe_get(f"{BASE}/{product_id}")

    if response is None:
        continue  # error already printed by safe_get

    if response.status_code == 404:
        print("Product not found.")
    elif response.status_code == 200:
        data = response.json()
        print(f"Product: {data['title']} — Price: ${data['price']}")
    else:
        print("Unexpected server response:", response.status_code)
```

```python
# Task 4 — Logging to File (Complete Solution)
# Step 1 — Create the Logger Function

from datetime import datetime

def log_error(source, error_type, message):
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    log_entry = f"[{timestamp}] ERROR in {source}: {error_type} — {message}\n"

    with open("error_log.txt", "a", encoding="utf-8") as log_file:
        log_file.write(log_entry)
```

```python
# Step 3 — Trigger an HTTP Error (404 Not Found)

try:
    response = requests.get("https://dummyjson.com/products/999", timeout=5)

    if response.status_code != 200:
        print("Simulated HTTP error caught.")
        log_error(
            source="lookup_product",
            error_type="HTTPError",
            message=f"{response.status_code} Not Found for product ID 999"
        )

except Exception as e:
    # This block normally won't run for 404, but included for safety
    log_error("lookup_product", "UnexpectedError", str(e))
```
